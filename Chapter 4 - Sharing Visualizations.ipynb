## 1. Changing plot style

### 1. 1 Dataset Description

The data consists of the weather information in Seattle and Austin for each month of the year. Columns include values such as monthy averages of temperature, precipitation, and their respective standard deviations. A link to the explanation of each of the columns is added in the Appendix.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df_seattle = pd.read_csv('data/seattle_weather.csv')
df_seattle = df_seattle[:12]
df_seattle["MONTHS"] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_austin = pd.read_csv('data/austin_weather.csv')
df_austin["MONTHS"] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'])
ax.plot(df_austin['MONTHS'],df_austin['MLY-TAVG-NORMAL'])
plt.show()
```

### 1.2 Plotting R-style using ggplot

```python
plt.style.use("ggplot")
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'])
ax.plot(df_austin['MONTHS'],df_austin['MLY-TAVG-NORMAL'])
ax.set_xlabel("Time(months)")
ax.set_ylabel("Average temperature (Fahrenheit)")
ax.set_title("Weather")
plt.show()
```

**Note:** To go back to the default style use `plt.style.use("default")`

### 1.3 Plotting bmh-style 

```python
plt.style.use("bmh")
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'])
ax.plot(df_austin['MONTHS'],df_austin['MLY-TAVG-NORMAL'])
ax.set_xlabel("Time(months)")
ax.set_ylabel("Average temperature (Fahrenheit)")
ax.set_title("Weather")
plt.show()
```

```python
plt.style.use("seaborn-colorblind")
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'])
ax.plot(df_austin['MONTHS'],df_austin['MLY-TAVG-NORMAL'])
ax.set_xlabel("Time(months)")
ax.set_ylabel("Average temperature (Fahrenheit)")
ax.set_title("Weather")
plt.show()
```

### 1.4 Plotting solarize-style 

```python
plt.style.use("Solarize_Light2")
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'])
ax.plot(df_austin['MONTHS'],df_austin['MLY-TAVG-NORMAL'])
ax.set_xlabel("Time(months)")
ax.set_ylabel("Average temperature (Fahrenheit)")
ax.set_title("Weather")
plt.show()
```

### Tips:
    
1. Use light backgrounds
2. Use a colorblind friendly options 
3. Use less ink if the figure is going to be printed, i.e avoid backgrounds
4. Use greyscale style if using a black and white printer

## 2. Saving your visualizations

Reading the Olympic data about medals won by different countries to use as an example. 

```python
medals = pd.read_csv('data/medals_by_country_2016.csv', )
#set the unnamed column to an index
medals.set_index('Unnamed: 0', inplace=True)
medals
```

### 2.1 Saving figure to file 

```python
plt.style.use("seaborn-colorblind")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
fig.savefig("images/gold_medals.png")
```

### 2.2 Different file formats

The `.png` format provides lossless compression of image and it retains high quality and have a larger file size. 

The `.jpg` format is recommended for use on websites as it is a lossy compression and can use to save images that use less disk space and bandwidth. 
The degree of loss of quality can be controlled by using the `quality` argument. 

The `.svg` format reuslts in a vector graphics file where different elements can be edited by advanced graphics software like Adobe Illustrator. 

The `dpi`, dots per inch is used to control the quality of the image. A value of `300` will give a good quality image. 

```python
plt.style.use("seaborn-colorblind")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
fig.savefig("images/gold_medals_300dpi.png", dpi=300)
```

### 2.3 Figure size
The size of the image determines the aspect ratio of the image result in a wide and short or a long and narrow image. It can be set using the `fig.set_size_inches()` command. 

```python
plt.style.use("seaborn-colorblind")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
fig.set_size_inches([3,5])
```

```python
plt.style.use("seaborn-colorblind")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
fig.set_size_inches([5,2])
```

## 3. Automating figures from data

Advantages of automating saving of files:
1. Ease and speed
2. Flexibility
3. Robustness
4. Reproducibility


Reading the data for automating figures from data: 

```python
summer_2016_medals = pd.read_csv('data/summer2016.csv')
sports = summer_2016_medals['Sport'].unique()
sports
```

```python
fig, ax = plt.subplots()

for sport in sports:
    sport_df = summer_2016_medals[summer_2016_medals['Sport'] == sport]
    ax.bar(sport, sport_df['Height'].mean(), yerr=sport_df['Height'].std())
ax.set_ylabel("Height (cm)")
ax.set_xticklabels(sports, rotation=90)
plt.tight_layout()
plt.show()
```
