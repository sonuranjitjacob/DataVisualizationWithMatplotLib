# Data Visualization with Matplotlib - Introduction

## 1. Dataset Description

The data consists of the weather information in Seattle and Austin for each month of the year. Columns include values such as monthy averages of temperature, precipitation, and their respective standard deviations. A link to the explanation of each of the columns is added in the Appendix. 

```python
import pandas as pd
import numpy as np 

df_seattle = pd.read_csv('data/seattle_weather.csv')
df_seattle = df_seattle[:12]
df_seattle["MONTHS"] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_seattle.head()
```

```python
df_austin = pd.read_csv('data/austin_weather.csv')
df_austin["MONTHS"] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_austin.head()
```

## 2.  Creating a figure

We create an empty template to plot the figure as follows.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.show()
```

Creating a figure with a plot:

We plot the `month` vs the `monthly average temperature`. 

```python
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'], marker='o')
plt.show()
```

## 3. Customizing data appearance

We can customize the above plot by varying the linestyle, marker and colour. We also add a title, x and y labels to the plot.

Changing the linestyle of the plot:

```python
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'], marker='o', linestyle="--")
plt.show()
```

Plotting only the points:

```python
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'], marker='o', linestyle="None")
plt.show()
```

Changing the color of the plot:

```python
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'], marker='o', linestyle="--", color='r')
plt.show()
```

Adding labels to the axes and a title to the plot:

```python
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-TAVG-NORMAL'], marker='o', linestyle="--", color='r')
ax.set_xlabel("Time(months)")
ax.set_ylabel("Average temperature (Fahrenheit)")
ax.set_title("Weather")
plt.show()
```

## 4. Creating multiple plots together:

To get a overview of the data, we may need to have multiple plots on the same figure. 

Here, we plot the monthly preicipitation with the 25th and 75th percentiles of daily nonzero precipitation totals for 29-day windows centered on each day of the year [1]. 

```python
fig, ax = plt.subplots()
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-PRCP-NORMAL'], marker='o', linestyle="-", color='b')
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-PRCP-25PCTL'], marker='o', linestyle="--", color='b')
ax.plot(df_seattle['MONTHS'],df_seattle['MLY-PRCP-75PCTL'], marker='o', linestyle="--", color='b')

ax.set_xlabel("Time (months)")
ax.set_ylabel("Precitipation (inches) ")
ax.set_title("Weather in Seattle")
plt.show()
```

## 5. Subplots

We can also create multiple figures together for visualization of different datasets. Creating a template for a subplot with three rows and two columns:

```python
fig, ax = plt.subplots(3, 2)
plt.show()
```

Creating two subplots one below the other where `ax[0]` refers to the first subplot and `ax[1]` to the second:

```python
fig, ax = plt.subplots(2,1, sharey=True)

ax[0].plot(df_seattle['MONTHS'],df_seattle['MLY-PRCP-NORMAL'], marker='o', linestyle="-", color='b')
ax[0].plot(df_seattle['MONTHS'],df_seattle['MLY-PRCP-25PCTL'], marker='o', linestyle="--", color='b')
ax[0].plot(df_seattle['MONTHS'],df_seattle['MLY-PRCP-75PCTL'], marker='o', linestyle="--", color='b')


ax[1].plot(df_austin['MONTHS'],df_austin['MLY-PRCP-NORMAL'], marker='o', linestyle="-", color='r')
ax[1].plot(df_austin['MONTHS'],df_austin['MLY-PRCP-25PCTL'], marker='o', linestyle="--", color='r')
ax[1].plot(df_austin['MONTHS'],df_austin['MLY-PRCP-75PCTL'], marker='o', linestyle="--", color='r')

ax[1].set_xlabel("Time (months)")
ax[0].set_ylabel("Precitipation (inches) ")
ax[1].set_ylabel("Precitipation (inches) ")

plt.show()
```

The `sharey` argument makes sure the two subplots have the same range for the y-axis. This makes comparing the data easier.

Creating subplots with each plot having different axes:

```python
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Addressing the top left Axes as index 0, 0, plot month and Seattle precipitation
ax[0, 0].plot(df_seattle['MONTHS'], df_seattle['MLY-PRCP-NORMAL'], marker='o', linestyle='--', color='r')

# In the top right (index 0,1), plot month and Seattle temperatures
ax[0, 1].plot(df_seattle['MONTHS'], df_seattle['MLY-TAVG-NORMAL'], marker='o', linestyle='--', color='r')

# In the bottom left (1, 0) plot month and Austin precipitations
ax[1,0].plot(df_seattle['MONTHS'], df_seattle['MLY-PRCP-NORMAL'], marker='o', linestyle='--', color='b')

# In the bottom right (1, 1) plot month and Austin temperatures
ax[1,1].plot(df_seattle['MONTHS'], df_seattle['MLY-TAVG-NORMAL'], marker='o', linestyle='--', color='b')

ax[0, 0].set_xlabel("Time (months)")
ax[0, 0].set_ylabel("Precitipation (inches) ")
ax[0, 1].set_ylabel("Temperature (inches) ")
ax[1, 0].set_ylabel("Precitipation (inches) ")
ax[1, 1].set_ylabel("Temperature (inches) ")

plt.tight_layout()
plt.show()
```
